```python
import torch
import torch.nn as nn
from torch.optim import Adam
from torch.utils.data import Dataset, DataLoader
from torchsummary import summary
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print (device)
```

```python
data_df = pd.read_csv("/content/rice-type-classification/riceClassification.csv")
data_df.head()
```

```python
data_df.dropna(inplace = True)
data_df.drop(('id'), axis = 1, inplace = True)
#shape returns np array or pd df as a tuple (rows, columns)
print(data_df.shape)
```

```python
data_df.head()
```

```python
print(data_df["Class"].unique())
```

```python
print(data_df['Class'].value_counts())
```

```python
original_df = data_df.copy()
#this is normalization to keep data ranging around -1 to 1
for column in data_df.columns:
  data_df[column] = data_df[column]/data_df[column].abs().max()

data_df.head()
```

```python
#iloc selects rows and columns by their integer position
#all rows, all columns but the last one
X = np.array(data_df.iloc[:, :-1])
#all rows and only the last column
Y = np.array(data_df.iloc[:, -1]) #np.array() converts data into a numpy array
```

```python
#training
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3) #30% for testing, 70% for training
```

```python
#testing
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5) # half half, test_size is a keyword
```
